# 2.什么叫过拟合，常用的减轻过拟合的方法有？

### 过拟合：学到了一些无用的东西，算法在训练集上表现好，但在测试集上表现不好，泛化性能差。

## 解决方法
    ·对数据的处理，数据不纯可能导致过拟合，我们可以重新清洗数据
    ·增大数据的训练量，过拟合的原因可能是因为训练集太小导致的
    ·正则化：Lasso回归和岭回归，即 L1正则化，直接删除高次项的特征，L2正则化，削弱特征，使高次项趋于0
    ·减少维度特征，防止维灾难

# 3.分类算法、聚类算法的区别是什么，有监督学习和无监督学习的区别是什么

分类算法：有特征值，有目标值，是有监督学习。分类可以是二分类，也可以是多分类。根据已知的一些样本（包括属性与类标号）来得到分类模型（即得到样本属性与类标号之间的函数），然后通过此目标函数来对只包含属性的样本数据进行分类

聚类算法：有特征值，无目标值，是无监督学习。聚类算法事先不知道目标值，一般通过计算组内和组间相似程度或者不相似程度来进行判别

有监督学习和无监督学习的区别：有监督学习有特征值和目标值，无监督学习有特征值无目标值

### libsvm的数据格式

 [label] [index1]:[value1]   [index2]:[value2] …

 [label] [index1]:[value1]   [index2]:[value2] …

```python
#     原始数据：
#              特征值a   特征值b  ...  目标值
#
#                 7         5     ...     2
#
#                 4         2     ...     1

#     转换之后：
#             目标值     条件属性a    特征值b  ...
#
#                 2        1:7          2:5    ...
#
#                 1       1:4           2:2    ...
```

```python
import pandas as pd
from sklearn import svm
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
# from libsvm.python.svmutil import *


# 导入数据
data = pd.DataFrame([[0.697, 0.460, 1],
        [0.774, 0.376, 1],
        [0.634, 0.264, 1],
        [0.608, 0.318, 1],
        [0.556, 0.215, 1],
        [0.430, 0.237, 1],
        [0.481, 0.149, 1],
        [0.437, 0.211, 1],
        [0.666, 0.091, 0],
        [0.243, 0.267, 0],
        [0.245, 0.057, 0],
        [0.343, 0.099, 0],
        [0.639, 0.161, 0],
        [0.657, 0.198, 0],
        [0.360, 0.370, 0],
        [0.593, 0.042, 0],
        [0.719, 0.103, 0]],columns = ['density', 'sugar_rate', 'label'])
train = data[["density", "sugar_rate"]]
test = data[["label"]]
data.head()
```

### 将数据转化为libsvm数据集

```python
x_train, x_test, y_train, y_test = train_test_split(train, test, test_size = 0.2, random_state = 666)
```

```python

def getX(X):             
    temp = []
    X = np.array(X)
    for i in range(np.shape(X)[0]):
        dic = {}
        for j in range(np.shape(X)[1]):
            dic[j+1] = X[i,j]
        temp.append(dic)
    return temp
train = getX(train)
test = getX(test)
train
```

```python
## 数据的训练
linear_svm = svm.SVC(C=0.5, #惩罚参数
            kernel='linear')
linear_svm.fit(x_train,y_train)
y_pred = linear_svm.predict(x_test)
print('linear_svm的准确率: ' , (accuracy_score(y_pred, y_test))) 
```
