```python
import pandas as pd
from sklearn import svm
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
# from libsvm.python.svmutil import *


# 导入数据
data = pd.DataFrame([[0.697, 0.460, 1],
        [0.774, 0.376, 1],
        [0.634, 0.264, 1],
        [0.608, 0.318, 1],
        [0.556, 0.215, 1],
        [0.430, 0.237, 1],
        [0.481, 0.149, 1],
        [0.437, 0.211, 1],
        [0.666, 0.091, 0],
        [0.243, 0.267, 0],
        [0.245, 0.057, 0],
        [0.343, 0.099, 0],
        [0.639, 0.161, 0],
        [0.657, 0.198, 0],
        [0.360, 0.370, 0],
        [0.593, 0.042, 0],
        [0.719, 0.103, 0]],columns = ['density', 'sugar_rate', 'label'])
train = data[["density", "sugar_rate"]]
test = data[["label"]]
```

```python
x_train, x_test, y_train, y_test = train_test_split(train, test, test_size = 0.4, random_state = 666)
## 数据的训练
linear_svm = svm.SVC(C=0.5, kernel='linear')   # C=, 惩罚参数
linear_svm.fit(x_train,y_train)
y_pred = linear_svm.predict(x_test)
print('linear_svm的准确率: ' , accuracy_score(y_pred, y_test))
```

# 3.手动或写代码计算输入是1，2，3时的网络输出。

```python
#!/usr/bin/env python
# -*- coding:utf-8 -*-
# author:杜培博
# datetime:2021/12/10 18:19
# software: PyCharm
import numpy as np
import pandas as pd
```

```python
# 数据的准备
np.random.seed(20211210)
zz = np.random.randn(1000)
w1 = [zz[0:4], zz[4:8], zz[8:12]]
w2 = [zz[12:15], zz[15:18], zz[18:21], zz[21:24]]
b = [zz[24:27], zz[27:31]]
train = [1, 2, 3]
train_fir = []
train_sec = []
```

```python
# sigmoid函数：
def sigmoid(x):
    res = []
    for i in x:
        res.append(1/(1+np.exp(-i)))
    return res
```

```python
# 第一层
c = 0
for i in w2:
    train_fir.append(np.dot(train, w2[c]) + b[1][c])
    c += 1
```

```python
# 第二层
tem = 0
for i in w1:
    train_sec.append(np.dot(train_fir, w1[tem]) + b[0][tem])
    tem += 1
```

```python
res = sigmoid(train_sec)
print("该神经网络的输出是：", res)
```
